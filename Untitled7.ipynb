```python
X = stock.drop(['Close','Name'], axis=1)
y = stock['Close']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics
```

```python
# Instead of using classification models, use regression models for continuous target variables.
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score


models = [LinearRegression(), SVR(kernel='poly'), XGBRegressor()]

for i in range(3):
    models[i].fit(X_train, y_train)

    print(f'{models[i]} : ')
    print('Training R^2 : ', r2_score(y_train, models[i].predict(X_train)))
    print('Validation R^2 : ', r2_score(y_test, models[i].predict(X_test)))
    print()
```

```python
# Iterate through each model in the list and make predictions
for model in models:
    predictions = model.predict(X_test)
    # Now you can use the 'predictions' variable for further analysis or evaluation
    print(f"Predictions using {model}: {predictions}") # Optionally print to view the output
```
